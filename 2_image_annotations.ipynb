```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

## Read Image

```python
pth = '/Users/gouher/Documents/personal/codes/computer_vision/image_processing_concepts/data/color/boat.jpg'
img = cv2.imread(pth,cv2.IMREAD_COLOR)
plt.imshow(img[:,:,::-1])
```

## Draw Line

```python
img_rgb = img[:,:,::-1]
img_line = img_rgb.copy()
img_line = cv2.line(img_line,pt1=(1000,1000),pt2=(2000,1000),color=(255,255,0),thickness=5,lineType=cv2.LINE_AA)
plt.imshow(img_line)
```

## Draw Circle

```python
img_circle = img_rgb.copy()
img_circle = cv2.circle(img_circle,center=(1100,1500),radius=500,color=(255,255,0),thickness=2,lineType=cv2.LINE_AA)
plt.imshow(img_circle)
```

## Draw Rectangle

```python
img_rect = img_rgb.copy()
img_rect = cv2.rectangle(img_rect,pt1=(800,1100),pt2=(1500,1900),color=(255,255,0),thickness=5,lineType=cv2.LINE_AA)
plt.imshow(img_rect)
```

## Put Text

```python
text = 'Boat in water'
fontScale = -7
fontFace = cv2.FONT_HERSHEY_PLAIN
fontColor = (0,255,0)
fontThickness = 5
org = (800,1000)
```

```python
img_txt = img_rect.copy()
img_txt = cv2.putText(img_txt,text=text,org=org,color=fontColor,fontScale=fontScale,thickness=fontThickness,fontFace=fontFace)
plt.imshow(img_txt)
```
