```python
img = cv2.imread(pt,cv2.IMREAD_COLOR)
print(img)
```

## Converting to different Color Spaces

```python
# cv2 reads in BGR format but plt expects in RGB, that's why the image shows blue
plt.imshow(img)
```

```python
# invert the channel order before passing to plt
type(plt.imshow(img[:,:,::-1]))
```

```python
# invert the channel to RGB order using cv2
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# invert the channel to HSV order using cv2
img_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)
```

```python
plt.imshow(cv2.cvtColor(img_rgb,cv2.COLOR_RGB2HSV))
```

## Splitting and Merging Color Channels

```python
h,s,v = cv2.split(img_hsv)
```

```python
img_merged = cv2.merge((h+10,s,v))
plt.imshow(img_merged)
```

```python
img_merged = cv2.cvtColor(img_merged,cv2.COLOR_HSV2RGB)
plt.imshow(img_merged)
```
