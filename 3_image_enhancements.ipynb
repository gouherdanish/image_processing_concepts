```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

## Sample Image

```python
op = '/Users/gouher/Documents/personal/codes/computer_vision/image_processing_concepts/data/color/scene_halved.png'
Image(op)
```

```python
img = cv2.imread(op, cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

## Brightness (Addition)

Change in brightness means changing the intensity of all the pixels by the same amount

This results in global increase/decrease in brightness

```python
img_rgb.shape
```

```python
A = np.ones(img_rgb.shape,dtype='uint8') * 50
A.shape
```

```python
img_bright = cv2.add(img_rgb, A)
img_dark = cv2.subtract(img_rgb, A)
```

```python
plt.figure(figsize=(18,5))
plt.subplot(131); plt.imshow(img_dark); plt.title('Darker')
plt.subplot(132); plt.imshow(img_rgb); plt.title('Original')
plt.subplot(133); plt.imshow(img_bright); plt.title('Brighter')
```

## Contrast (Multiplication)

Contrast is the difference in intensity values of pixels.

Changing contrast means increasing or decresing this difference and this can be achieved by multiplication

```python
A1 = np.ones(img_rgb.shape) * .8
A2 = np.ones(img_rgb.shape) * 1.2
```

```python
img_darker = np.uint8(cv2.multiply(np.float64(img_rgb),A1))
img_brighter = np.uint8(cv2.multiply(np.float64(img_rgb),A2))
```

```python
plt.figure(figsize=(18,5))
plt.subplot(131); plt.imshow(img_darker); plt.title('Lower Contrast')
plt.subplot(132); plt.imshow(img_rgb); plt.title('Original')
plt.subplot(133); plt.imshow(img_brighter); plt.title('Higher Contrast')
```

```python
img_darker = np.uint8(cv2.multiply(np.float64(img_rgb),A1))
img_brighter = np.uint8(np.clip(cv2.multiply(np.float64(img_rgb),A2),0,255))
```

```python
plt.figure(figsize=(18,5))
plt.subplot(131); plt.imshow(img_darker); plt.title('Lower Contrast')
plt.subplot(132); plt.imshow(img_rgb); plt.title('Original')
plt.subplot(133); plt.imshow(img_brighter); plt.title('Higher Contrast')
```

## Bitwise Operations

```python
p1 = '/Users/gouher/Documents/personal/codes/computer_vision/image_processing_concepts/data/gray/circle.png'
p2 = '/Users/gouher/Documents/personal/codes/computer_vision/image_processing_concepts/data/gray/rectangle.png'

img_cir = cv2.imread(p1, cv2.IMREAD_GRAYSCALE)
img_rec = cv2.imread(p2, cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(20,5))
plt.subplot(121); plt.imshow(img_rec,cmap='gray'); plt.title('Rectangle')
plt.subplot(122); plt.imshow(img_cir,cmap='gray'); plt.title('Circle')
```

```python
img_rec.shape, img_cir.shape
```

```python
# Should be of same size for bitwise ops, so resizing
img_rec = cv2.resize(img_rec,dsize=(499,200))
img_cir = cv2.resize(img_cir,dsize=(499,200))
```

```python
img_rec.shape
```

### 1. Bitwise AND

```python
img_and = cv2.bitwise_and(img_rec,img_cir,mask=None)
plt.imshow(img_and,cmap='gray')
```

### 2. Bitwise OR

```python
img_and = cv2.bitwise_or(img_rec,img_cir,mask=None)
plt.imshow(img_and,cmap='gray')
```

### 3. Bitwise XOR

_XOR output is High(1) only when exactly one of the input is High(1)_

```python
img_and = cv2.bitwise_xor(img_rec,img_cir,mask=None)
plt.imshow(img_and,cmap='gray')
```
