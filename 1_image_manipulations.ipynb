## Creating Image

```python
# Checkerboard Sample
arr = [
        [0,0,0,255,255,255,0,0,0],
        [0,0,0,255,255,255,0,0,0],
        [0,0,0,255,255,255,0,0,0],
        [255,255,255,0,0,0,255,255,255],
        [255,255,255,0,0,0,255,255,255],
        [255,255,255,0,0,0,255,255,255],
        [0,0,0,255,255,255,0,0,0],
        [0,0,0,255,255,255,0,0,0],
        [0,0,0,255,255,255,0,0,0]
    ]
npa = np.array(arr)
```

```python
plt.imshow(npa,cmap='gray')
```

## Cropping Image

```python
pth = '/Users/gouher/Documents/personal/codes/computer_vision/image_processing_concepts/data/color/boat.jpg'
img = cv2.imread(pth,cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_cropped = img_rgb[1100:1800,800:1500]
plt.imshow(img_cropped)
```

```python
img_rgb.shape, img_cropped.shape
```

## Resizing Image
_Keeps the Image content and sanctity same and resizes by changing the number of pixels, so the image size changes_

### 1. Scale image by factor of 2 (Doubles the number of pixels)

```python
img_resized = cv2.resize(img_cropped, None, fx=2,fy=2)
plt.imshow(img_resized)
```

### 2. Set specific size to the image
_Image gets distorted_

```python
img_res2 = cv2.resize(img_cropped,dsize=(2000,1000))
plt.imshow(img_res2)
```

```python
img_res2 = cv2.resize(img_cropped,dsize=(2000,1000),interpolation=cv2.INTER_AREA)
plt.imshow(img_res2)
```

### 3. Resize by maintaining aspect ratio

```python
aspect_ratio = img_cropped.shape[0] / img_cropped.shape[1]
aspect_ratio
```

```python
reqd_width = 1000
req_height = int(aspect_ratio * reqd_width)
req_height
```

```python
img_res3 = cv2.resize(img_cropped,dsize=(reqd_width,req_height),interpolation=cv2.INTER_AREA)
plt.imshow(img_res3)
```

## Rotating/Flipping Image

### 1. Horizontal Flip (around y-axis)

```python
img_flip1 = cv2.flip(img_cropped,flipCode=1)
plt.imshow(img_flip1)
```

### 2. Vertical Flip (around x-axis)

```python
img_flip1 = cv2.flip(img_cropped,flipCode=0)
plt.imshow(img_flip1)
```

### 3. Both Axes Flip

```python
img_flip1 = cv2.flip(img_cropped,flipCode=-1)
plt.imshow(img_flip1)
```
