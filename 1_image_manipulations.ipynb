```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from IPython.display import Image
```

## Creating Image

```python
# Checkerboard Sample
arr = [
        [0,0,0,255,255,255,0,0,0],
        [0,0,0,255,255,255,0,0,0],
        [0,0,0,255,255,255,0,0,0],
        [255,255,255,0,0,0,255,255,255],
        [255,255,255,0,0,0,255,255,255],
        [255,255,255,0,0,0,255,255,255],
        [0,0,0,255,255,255,0,0,0],
        [0,0,0,255,255,255,0,0,0],
        [0,0,0,255,255,255,0,0,0]
    ]
npa = np.array(arr)
```

```python
plt.imshow(npa,cmap='gray')
```

## Modifying Image

```python
npa_mod = npa.copy()
npa_mod[1,1] = 200
npa_mod[7,1] = 200
npa_mod[4,4] = 200
npa_mod[1,7] = 200
npa_mod[7,7] = 200
npa_mod
```

```python
plt.imshow(npa_mod,cmap='gray')
```
