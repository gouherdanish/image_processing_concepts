```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

## Sample Image

```python
pt = '/Users/gouher/Documents/personal/codes/computer_vision/image_processing_concepts/data/gray/piano_music_sheet.png'
img = cv2.imread(pt,cv2.IMREAD_GRAYSCALE)
plt.imshow(img,cmap='gray')
```

## Global Thresholding - 1

Using a lower threshold misses some of the notes

```python
retval, img_thresh = cv2.threshold(img, thresh=50, maxval=255, type=cv2.THRESH_BINARY)
plt.imshow(img_thresh,cmap='gray')
```

## Global Thresholding - 2

Using a higher threshold includes black areas which are not actually notes

```python
retval2, img_thresh2 = cv2.threshold(img, thresh=130, maxval=255, type=cv2.THRESH_BINARY)
plt.imshow(img_thresh2,cmap='gray')
```

## Adaptive Thresholding

```python
img_adapted = cv2.adaptiveThreshold(img, maxValue=255, adaptiveMethod=cv2.ADAPTIVE_THRESH_MEAN_C, thresholdType=cv2.THRESH_BINARY, blockSize=11, C=7)
plt.imshow(img_adapted,cmap='gray')
```
